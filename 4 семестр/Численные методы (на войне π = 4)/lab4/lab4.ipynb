```python
import numpy as np

def symmetric_derivative(f, x0, epsilon):
    h = np.sqrt(epsilon)
    central_derivative = (f(x0 + h) - f(x0 - h)) / (2 * h)
    return central_derivative

real_derivative = lambda x: np.exp(x/3) / 3

f = lambda x: np.exp(x/3)
x0 = 0.9

epsilon1 = 10**(-3)
epsilon2 = 10**(-6)

diff1 = symmetric_derivative(f, x0, epsilon1)
diff2 = symmetric_derivative(f, x0, epsilon2)

print("Формула средней разностной производной:")
print(f"ε = {epsilon1}   f`({x0}) = {diff1}   Δ = {abs(diff1 - real_derivative(x0))}")
print(f"ε = {epsilon2}   f`({x0}) = {diff2}   Δ = {abs(diff2 - real_derivative(x0))}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def classic_derivative(f, x0, h):
    return (f(x0 + h) - f(x0)) / h

def symmetric_derivative(f, x0, h):
    return (f(x0 + h) - f(x0 - h)) / (2 * h)

real_derivative = lambda x: np.exp(x/3) / 3
f = lambda x: np.exp(x/3)
x0 = 0.9

steps = [1/2**i for i in range(1, 10)]

classic_derivative_error = [round(abs(classic_derivative(f, x0, h) - real_derivative(x0)), 14) for h in steps]
symmetric_derivative_error = [round(abs(symmetric_derivative(f, x0, h) - real_derivative(x0)), 14) for h in steps]

plt.grid()
plt.plot(steps, classic_derivative_error, marker="o",   label="Классическая формула")
plt.plot(steps, symmetric_derivative_error, marker="o", label="Средняя разностная ф-ла")
plt.xlabel("Шаг h")
plt.ylabel("Абсолютная погрешность Δ")
plt.legend()
plt.title("Ошибки методов дифференцирования")
```

```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.exp(x/3)
x0 = 0.9
real_derivative = lambda x: np.exp(x/3)/3

steps = [1/2**i for i in range(50)] 

errors = [abs((f(x0 + h) - f(x0))/h - real_derivative(x0)) for h in steps]

plt.loglog(steps, errors, marker="o")
plt.grid()
plt.xlabel("Шаг h")
plt.ylabel("Абсолютная погрешность Δ")
plt.title("Неустойчивость численного дифференцирования (правосторонняя разность)")
plt.show()
```

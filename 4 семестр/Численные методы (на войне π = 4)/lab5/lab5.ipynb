```python
import numpy as np

def f(x):
    return 1 / (1 + x**2)

a, b = 0, 1

M4 = 24 # Максимум 4-й производной

eps_pi = 1e-6

# I = π/4
eps = eps_pi / 4

# Из оценки погрешности Симпсона:
# Ψ ≤ (b-a)/2880 * h^4 * M4
h = ((2880 * eps) / ((b - a) * M4)) ** 0.25

# Делаем n таким, чтобы было целое кол-во отрезков на [a, b]
n = int(np.ceil((b - a) / h))
h = (b - a) / n

x = np.linspace(a, b, n + 1)

s1 = sum(f(x[i]) for i in range(1, n, 2))
s2 = sum(f(x[i]) for i in range(2, n, 2))

I = h/3 * (f(a) + f(b) + 4*s1 + 2*s2)

pi_approx = 4 * I

print(f"n = {n}")
print(f"h = {h}")
print(f"π ≈ {pi_approx}")
print(f"δ = {abs(pi_approx-np.pi)/np.pi}")
```

```python
import numpy as np

def f(x):
    return 1 / (1 + x**2)

a, b = 0, 1
eps = 1e-6

def simpson(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)

    s1 = sum(f(x[i]) for i in range(1, n, 2))
    s2 = sum(f(x[i]) for i in range(2, n, 2))

    return h / 3 * (f(a) + f(b) + 4 * s1 + 2 * s2)


n = 4
I_h = simpson(f, a, b, n)

while True:
    n *= 2
    I_h2 = simpson(f, a, b, n)

    if abs(I_h2 - I_h) / 15 < eps:
        break
        
    I_h = I_h2

pi_approx = 4 * I_h2

print(f"n = {n}")
print(f"π ≈ {pi_approx}")
print(f"δ = {abs(pi_approx-np.pi)/np.pi}")
```
